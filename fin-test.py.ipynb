```python
stock_df_long = pd.melt(stock_df, value_vars=[ 'Close', 'close_30day', 'close_90day']
        , ignore_index=False, var_name='metric', value_name='value')
```

```python
import plotly.express as px

fig = px.line(stock_df_long,y="value", color="metric", title='Close of JPM')
fig.show()
```

```python
from ipywidgets import DatePicker, FloatText,HBox, interactive_output, Text
import datetime
from dateutil.relativedelta import relativedelta
import numpy as np

date_end = DatePicker(value=datetime.date.today())
date_start = DatePicker(value=datetime.date.today() + relativedelta(years=-1))
symbol = widgets.Text(value='TSLA')

ui = HBox([date_start, date_end, symbol])

def get_ticker(sym):
    if not sym:
        sym = "JPM"
    return yf.Ticker(sym)


global ticker 
ticker = get_ticker(symbol.value)   
```

```python
def f(end, start, sym):
    ticker = get_ticker( sym )

    tsla_df = ticker.history(start=start, end=end)
    short_name = ticker.info['shortName']
    title = short_name + "'s stock price"
    tsla_df['Close'].plot(title=title)
    ticker.info['longBusinessSummary']

def i(sym):    
    #clear_output()
    if not sym:
        sym = "JPM"
    ticker = yf.Ticker(sym)
    display( ticker.info['longBusinessSummary'] )

out = interactive_output(f, {'end': date_end, 'start': date_start, 'sym': symbol})
out2 = interactive_output(i, {'sym': symbol})

display(ui, out, out2)
```
